### Question 1

Using the spreadsheet with the 2018 data, create a scatter plot that plots median family income against the percentage of the population with health insurance.  Change the color of the bubbles to red and use the documentation on matplotlib.pyplot.scatter to change the bubbles to crosses (5 points).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df = pd.read_csv("C:/Users/ankit/Downloads/2018 Data-1.csv")
```

```python
df.head()
```

```python
plt.scatter(df['Median Family Income'], df['Perc with health Insurance'], color='red', marker='x')

plt.xlabel('Median Family Income')
plt.ylabel('Percentage with Health Insurance')
plt.title('Scatter Plot: Median Family Income vs. Percentage with Health Insurance')


# Show plot
plt.grid(True)
plt.show()
```
