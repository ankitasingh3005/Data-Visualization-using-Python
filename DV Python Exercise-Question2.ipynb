### Question 2

Using the spreadsheet with the consolidated COVID data that you can download from the page A Line Chart in Python in this module, create a line graph that shows the moving average and the number of daily cases for Pasco

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df1 = pd.read_csv("C:/Users/ankit/Downloads/Pasco.csv")
```

```python
df1.head()
```

```python
df1 = df1.rename(columns={'Unnamed: 0': 'Date'})
df1 = df1.rename(columns={'Count of ObjectId (Pasco.csv)': 'Daily Cases'})
df1 = df1.rename(columns={'Moving Average Pasco along Day of EventDate (Pasco.csv)': 'Moving Average'})
```

```python
df1.head()
```

```python
df1.dtypes
```

```python
df1.plot('Date',['Daily Cases','Moving Average'])
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.title('Daily Cases and Moving Average for Pasco')
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
```
