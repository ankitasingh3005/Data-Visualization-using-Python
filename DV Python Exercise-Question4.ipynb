### Question 4
Using the simplified Car Sales Data, create a (horizontal) bar chart (not a column chart) that shows the total sales and total costs for the United Kingdom, the United States, and France

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df3 = pd.read_csv('Simplified Car Sales Data.csv')
```

```python
df3.head()
```

```python
df_filtered = df3[df3['CountryName'].isin(['United Kingdom', 'USA', 'France'])]

totals = df_filtered.groupby('CountryName').agg({'SalePrice': 'sum', 'CostPrice': 'sum'})
# Plot horizontal bar chart
totals.plot(kind='barh', figsize=(10, 6))

# Set labels and title
plt.xlabel('Amount')
plt.ylabel('Country')
plt.title('Total Sales and Costs by Country')

# Show plot
plt.show()
```
